```python
import pandas as pd

df = pd.read_csv("matches.csv")

# Keep only useful columns
df = df[['team1', 'team2', 'toss_winner', 'venue', 'winner']]

# Remove missing values
df = df.dropna()

print(df.head())
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['team1'] = le.fit_transform(df['team1'])
df['team2'] = le.fit_transform(df['team2'])
df['toss_winner'] = le.fit_transform(df['toss_winner'])
df['venue'] = le.fit_transform(df['venue'])
df['winner'] = le.fit_transform(df['winner'])
```

```python
X = df[['team1', 'team2', 'toss_winner', 'venue']]
y = df['winner']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)

print("Model trained successfully ✅")
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
import pandas as pd

df = pd.read_csv("matches.csv")

df = df[['team1', 'team2', 'toss_winner', 'venue', 'winner']]
df = df.dropna()
```

```python
df.tail()
```

```python
from sklearn.preprocessing import LabelEncoder

team_encoder = LabelEncoder()

# Combine ALL team-related columns (still TEXT)
all_teams = pd.concat([
    df['team1'],
    df['team2'],
    df['toss_winner'],
    df['winner']
])

team_encoder.fit(all_teams)

print(team_encoder.classes_)  # ✅ should show team names
```

```python
df['team1'] = team_encoder.transform(df['team1'])
df['team2'] = team_encoder.transform(df['team2'])
df['toss_winner'] = team_encoder.transform(df['toss_winner'])
df['winner'] = team_encoder.transform(df['winner'])
```

```python
venue_encoder = LabelEncoder()
df['venue'] = venue_encoder.fit_transform(df['venue'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X = df[['team1', 'team2', 'toss_winner', 'venue']]
y = df['winner']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier()
model.fit(X_train, y_train)

print("Model trained ✅")
```

```python
team1 = "Mumbai Indians"
team2 = "Chennai Super Kings"
toss = "Mumbai Indians"
venue = "Wankhede Stadium"

input_data = [[
    team_encoder.transform([team1])[0],
    team_encoder.transform([team2])[0],
    team_encoder.transform([toss])[0],
    venue_encoder.transform([venue])[0]
]]

prediction = model.predict(input_data)
print("Predicted winner:", team_encoder.inverse_transform(prediction))
```

```python
probs = model.predict_proba(input_data)

print(probs)
```

```python
classes = model.classes_

for i, prob in enumerate(probs[0]):
    team_name = team_encoder.inverse_transform([classes[i]])[0]
    print(f"{team_name} → {round(prob*100, 2)}%")
```

```python
team1 = "Mumbai Indians"
team2 = "Chennai Super Kings"

for i, prob in enumerate(probs[0]):
    team_name = team_encoder.inverse_transform([classes[i]])[0]
    
    if team_name == team1 or team_name == team2:
        print(f"{team_name} → {round(prob*100, 2)}%")
```

```python
winner = model.predict(input_data)
print("Predicted Winner:", team_encoder.inverse_transform(winner)[0])
```

```python
# Calculate win rate
win_counts = df['winner'].value_counts()

df['team1_win_rate'] = df['team1'].map(win_counts) / len(df)
df['team2_win_rate'] = df['team2'].map(win_counts) / len(df)
```

```python
def head_to_head(row):
    matches = df[(df['team1'] == row['team1']) & (df['team2'] == row['team2'])]
    return len(matches)

df['h2h'] = df.apply(head_to_head, axis=1)
```
